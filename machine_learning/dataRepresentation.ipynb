```python
# 数据表达 
# 使用哑变量转化类型特征
# 哑变量（Dummy Variables），也称为了虚拟变量，它能将某些类型变量的数据转化为二值变量，从而便于处理。

# 导入pandas
import pandas as pd
# 输入一个数据表
fruits = pd.DataFrame({'data features': [1, 2, 3, 4, 5],
                       'type features': ['Apple', 'Banana', 'Cherry', 'Durian', 'Eggplant']})
display(fruits)

# 转变数据表中的字符串为数值
fruits_dum = pd.get_dummies(fruits)
# 显示转变之后的数据表
display(fruits_dum)

# 如果想要将数值特征也进行get_dummies转换，可以先将数值特征转换为字符串，然后通过get_dummies的columns参数来转换
# 令程序将数值也看作字符串，实际上不转换数值为字符串也可以得到同样的结果，但为了避免错误最好是执行这一步
fruits['data features'] = fruits['data features'].astype(str)
# 使用get_dummies转换字符串
pd.get_dummies(fruits, columns=['data features'])
```

```python
# 对数据进行装箱处理
# 不同的算法建立的模型会有很大的差别，即使在同一数据集，这种差别也会存在，这是因为算法的工作原理不同导致的。
# 下面对比KNN和MLP这两种算法

# 导入numpy
import numpy as np
# 导入画图工具
import matplotlib.pyplot as plt
# 生成随机数列
rnd = np.random.RandomState(38)
x = rnd.uniform(-5, 5, size=50)
# 向数据集中添加噪声
y_no_noise = (np.cos(5*x)+x)
X = x.reshape(-1, 1)
y = (y_no_noise + rnd.normal(size=len(x)))/2

plt.plot(X, y, 'o', c='r')
plt.show()
```

```python
# 使用MLP算法和KNN算法分别对数据集进行回归分析

# 导入神经网络
from sklearn.neural_network import MLPRegressor
# 导入KNN
from sklearn.neighbors import KNeighborsRegressor
# 生成等差数列
line = np.linspace(-5, 5, 1000, endpoint=False).reshape(-1, 1)
# 分别用两种算法对其进行拟合
mlpr = MLPRegressor().fit(X, y)
knr = KNeighborsRegressor().fit(X, y)
# 绘制图形
plt.plot(line, mlpr.predict(line), label='MLP')
plt.plot(line, knr.predict(line), label='KNN')
plt.plot(X, y, 'o', c='r')
plt.legend(loc='best')
plt.show()
```

```python
# 装箱处理（binning），也成为了离散化处理（discretization）
# 设置箱体数为11
bins = np.linspace(-5, 5, 11)
# 对数据进行装箱处理
target_bin = np.digitize(X, bins=bins)
# 打印数据集形态
print('data shape before bin:{}\n'.format(target_bin.shape))
# 打印装箱的数据范围
print('bin;s data range:\n{}\n'.format(bins))
# 打印前十个数据的特征值
print('top 10 data features:\n{}\n'.format(X[:10]))
# 找到他们所在的箱子
print('top 10 data\'s bin:\n{}\n'.format(target_bin[:10]))
# 每两个整数值之间的区域就是一个箱子（或者说容器，container）
```

```python
# 将数据装箱后，需要做的事就是使用新的方法来表达它们。
# 这里使用sklearn的独热编码方法OneHotEncoder，它和pandas的get_dummies功能基本一样，但他只能处理整型数值的数据。

# 导入独热编码
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder(sparse=False)
onehot.fit(target_bin)
# 使用独热编码转化数据
X_in_bin = onehot.transform(target_bin)
# 打印结果
print('data shape after bin:{}\n'.format(X_in_bin.shape))
print('top 10 data:\n{}\n'.format(X_in_bin[:10]))

# 可以看到数据集样本数仍是50个，但特征数已变成10个，这是因为我们生成了10个箱子，而新的数据特征使用其所在的箱子号码来表示的。
# 这样一来，我们就把数据集中的连续特征转换为类别特征
```

```python
# 使用独热编码进行数据表达
new_line = onehot.transform(np.digitize(line, bins=bins))
# 使用新的数据来训练模型
new_mlpr = MLPRegressor().fit(X_in_bin, y)
new_knr = KNeighborsRegressor().fit(X_in_bin, y)
# 绘制图形
plt.plot(line, new_mlpr.predict(new_line), label='New MLP')
plt.plot(line, new_knr.predict(new_line), label='New KNN')
plt.plot(X, y, 'o', c='r')  # 原始数据
# 设置图注
plt.legend(loc='best')
# 显示图片
plt.show()

# 在这里也要对预测的数据进行相同的装箱处理，才能得到正确的预测结果。
# 可以看到数据集经过装箱处理后，这两个模型变得更加相似了，尤其是在x>0的部分。
# 从而可以看出对样本特征装箱的一个好处，那就是它可以纠正模型过拟合或者欠拟合的问题，
# 尤其是针对大规模高纬度的数据集使用线性模型的时候，装箱处理可以大幅提升线性模型的预测准确性。
```
